# Question 8

Monte Carlo simulations involve using randomness to solve problems that might be deterministic in principle. The idea for estimating $\pi$ with Monte Carlo simulation involves generating random points within a square and determining how many fall within a quarter circle inscribed within that square. The ratio of the points inside the circle to the total number of points, multiplied by $4$, gives an approximation of $\pi$.


1. **Conceptual Overview:**
   - Imagine a square with sides of length $2$, centered at the origin $(0,0)$, which means each side extends from $-1$ to $+1$. Inside this square, a circle with a radius of $1$ is inscribed.
   - The area of the circle is $\pi \times r^2 = \pi$, since $r = 1$, and the area of the square is $2 \times 2 = 4$.
   - The ratio of the area of the circle to the area of the square is $\frac{\pi}{4}$.
   - By randomly generating points within the square and counting how many fall inside the quarter circle (the circle's quadrant that's in the positive $x$ and $y$ space), we can estimate $\pi$.

2. **Procedure:**
   - Generate two random numbers between $-1$ and $1$. These represent the $x$ and $y$ coordinates of a point.
   - Determine whether this point is inside the quarter circle by checking if $x^2 + y^2 \leq 1$.
   - Repeat this process for a large number of points, and keep track of how many fall inside the quarter circle.
   - The ratio of points inside the quarter circle to the total number of points, multiplied by $4$, is an approximation of $\pi$.

3. **Implementation:**
   - Choose a large number of points, for example, $10,000$ or more, to get a better approximation.

let's implement this Monte Carlo simulation


This code simulates random points in a square that bounds a circle. By calculating the proportion of points that fall inside the circle compared to the total number of points in the square, we can approximate the value of π.

This code performs the following steps:
1. It defines a function `estimate_pi` that takes the number of points to simulate as input.
2. Within the function, it generates `num_points` random points within a square defined by x and y coordinates ranging from -1 to 1.
3. It then calculates the distance of each point from the origin (0,0) and determines whether the point is inside the circle inscribed in the square.
4. By comparing the number of points inside the circle to the total number of points, it estimates the value of π, leveraging the geometric relationship between the area of the circle and the square.
5. It runs this simulation for different numbers of points to see how the accuracy of the π estimate improves with more points.

```r
set.seed(123) # Setting a seed for reproducibility

# Function to estimate Pi using Monte Carlo simulation
estimate_pi <- function(num_points) {
  # Generate random points
  x <- runif(num_points, min=-1, max=1)
  y <- runif(num_points, min=-1, max=1)

  # Calculate the distance from (0, 0) and check if it's within the circle
  inside_circle <- sqrt(x^2 + y^2) <= 1

  # Estimate Pi
  estimated_pi <- (sum(inside_circle) / num_points) * 4

  return(estimated_pi)
}

# Number of points to simulate
points_to_simulate <- c(100, 1000, 10000, 100000, 1000000)

# Estimate Pi for different numbers of points
pi_estimates <- sapply(points_to_simulate, estimate_pi)

# Print the estimates
print(pi_estimates)
```

Based on the output, we can observe the estimated values of π for different numbers of points:

- For 100 points, the estimate is 3.40.
- For 1,000 points, the estimate is 3.176.
- For 10,000 points, the estimate is 3.1564.
- For 100,000 points, the estimate is 3.1436.
- For 1,000,000 points, the estimate is 3.142604.

These results illustrate a fundamental characteristic of Monte Carlo simulations: as the number of random points increases, the accuracy of the estimate improves. Initially, with a small number of points, the estimate of π may be quite far from the actual value. However, as we increase the number of points, the estimates become progressively closer to the true value of π (approximately 3.14159).
