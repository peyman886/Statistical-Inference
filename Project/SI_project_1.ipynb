# 4. Correlation and Visualization of Multiple Variables

### Feature Engineers
To enhance analysis, I created additional numeric features from the available categorical and continuous variables in the dataset. Specifically, I derived the height and BMI features based on the weight, age, race, and gender.

I create a manual model to predict the height, we can use the features `weight`, `ppage`, `race`, and `gender` to derive an estimated height. While this won't be as accurate as a model trained on actual height data, we can create a heuristic based on typical height distributions.

Here's a rough heuristic model:

1. **Gender:** Typically, males are taller than females.
2. **Age (ppage):** Heights can vary by age groups, but we'll use it mainly to ensure heights are within reasonable adult ranges.
3. **Race:** There are average height differences among different racial groups.
4. **Weight:** There is some correlation between height and weight, though it's not direct.

Let's define a basic formula to predict height. For simplicity, we'll assume:

- Base height for females is 160 cm, and for males is 170 cm.
- We adjust height slightly based on weight and age, with small modifications for race.

We'll create the height feature based on these assumptions.

```plaintext
height = base_height + weight_factor * weight + age_factor * (age - 50) + race_factor
```

To create the BMI (Body Mass Index) feature, we'll use the estimated height and weight. The BMI formula is:

$$ \text{BMI} = \frac{\text{weight (kg)}}{\left(\text{height (m)}\right)^2} $$

Since the height is estimated in centimeters, we need to convert it to meters before applying the formula.

````python
data=dataset
# Define base heights
base_height_male = 170  # cm
base_height_female = 160  # cm

# Define factors
weight_factor = 0.2  # cm per kg
age_factor = -0.05  # cm per year over 50
race_factors = {
    'White': 0,
    'Black': -2,
    'Hispanic': -1,
    'Asian': -3,
    'Other': 0
}

# Function to estimate height
def estimate_height(row):
    base_height = base_height_male if row['gender'] == 'Male' else base_height_female
    race_factor = race_factors.get(row['race'], 0)
    weight_adjustment = weight_factor * row['weight']
    age_adjustment = age_factor * (row['ppage'] - 50)
    return base_height + weight_adjustment + age_adjustment + race_factor

# Apply the function to the dataset
data['estimated_height'] = data.apply(estimate_height, axis=1)


# Convert estimated height from cm to meters
data['estimated_height_m'] = data['estimated_height'] / 100

# Calculate BMI using the formula
data['BMI'] = data['weight'] / (data['estimated_height_m'] ** 2)

# Drop the intermediate height in meters column
data.drop(columns=['estimated_height_m'], inplace=True)

# Save the modified dataset
modified_file_path = 'modified_nonvoters_data.csv'
data.to_csv(modified_file_path, index=False)

# Display the first few rows of the modified dataset
data.head()
````

## A. Heatmap correlogram with Pearson correlation coefficients and p-values.

We'll select 'weight', 'ppage', 'race', 'gender', 'estimated_height' and 'BMI' for demonstration.

````python
from scipy.stats import pearsonr

# Select numerical variables for the correlation matrix
numerical_vars = ['ppage', 'weight', 'estimated_height', 'BMI']

corr_matrix = dataset[numerical_vars].corr()

# Create a function to calculate the p-values
def calculate_pvalues(df):
    pvalues = pd.DataFrame(np.ones((df.shape[1], df.shape[1])), columns=df.columns, index=df.columns)
    for col in df.columns:
        for row in df.columns:
            pvalues.loc[row, col] = pearsonr(df[row], df[col])[1]
    return pvalues

# Calculate p-values
pvalues = calculate_pvalues(dataset[numerical_vars])

# Annotate the heatmap with correlation coefficients and p-values
annot = np.empty_like(corr_matrix, dtype=object)
for i in range(corr_matrix.shape[0]):
    for j in range(corr_matrix.shape[1]):
        annot[i, j] = f'{corr_matrix.iloc[i, j]:.2f}\n({pvalues.iloc[i, j]:.2e})'

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=annot, fmt='', cmap='coolwarm', center=0, linewidths=.5, cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Heatmap Correlogram with Correlation Coefficients and P-values')
plt.show()
````

# 6. Hypothesis Testing for Two Numerical Variables

## A. Best Method for Testing the Hypothesis:

The best method for testing our hypothesis is a **t-test**. The t-test is appropriate when comparing the means of two small samples (typically less than 30) and the population standard deviations are unknown. In this case, since we are using a random sample of 25 data points, the t-test is suitable.

## B. Hypothesis Test and Confidence Interval:

We use "ppage" (age) and "weight" for this analysis.


1. **Null Hypothesis (H0)**: The means of the two variables "ppage" (age) and "weight" are equal.
2. **Alternative Hypothesis (H1)**: The means of the two variables "ppage" (age) and "weight" are not equal.


````python
import numpy as np
from scipy.stats import ttest_ind
import random

# Set a seed for reproducibility
np.random.seed(42)

# Draw a random sample of 25 data points
sample_indices = np.random.choice(dataset.index, size=25, replace=False)
sample = dataset.loc[sample_indices]

# Select the numerical variables "ppage" (age) and "weight"
variable1 = sample['ppage']
variable2 = sample['weight']

# Perform a two-sample t-test
t_statistic, p_value = ttest_ind(variable1, variable2)

# Calculate the 95% confidence interval for the difference in means
mean_diff = np.mean(variable1) - np.mean(variable2)
std_err_diff = np.sqrt(np.var(variable1)/len(variable1) + np.var(variable2)/len(variable2))
confidence_interval = (mean_diff - 1.96 * std_err_diff, mean_diff + 1.96 * std_err_diff)

t_statistic, p_value, confidence_interval
````

**Results of the t-test:**
- **t-statistic**: 13.72
- **p-value**: 3.14e-18

Since the p-value is extremely small (much less than 0.05), we reject the null hypothesis. This indicates that there is significant evidence to suggest a difference in the mean values of "ppage" (age) and "weight".

**95% Confidence Interval for the Difference in Means:**
- Confidence Interval: (42.44, 56.26)

The 95% confidence interval does not include zero, which further supports the conclusion that there is a significant difference in the means of the two variables.

**Conclusion:**
The hypothesis test results agree with the 95% confidence interval. Both the t-test and the confidence interval provide convincing evidence of a difference between the mean values of "ppage" (age) and "weight".


# 7. Confidence Interval and Hypothesis Testing for a Numerical Variable

choose "ppage" (age)

## A. Confidence interval

````python
from scipy.stats import norm

# Sample mean and standard error
sample_mean = np.mean(dataset['ppage'])
sample_std = np.std(dataset['ppage'], ddof=1)
n = len(dataset['ppage'])
std_err = sample_std / np.sqrt(n)

# 98% confidence interval
conf_interval = norm.interval(0.98, loc=sample_mean, scale=std_err)
sample_mean, conf_interval
````

- **Sample Mean**: 51.69
- **98% Confidence Interval**: (51.17, 52.21)

## B. Interpretation of the confidence interval

In this context, we are 98% confident that the true population mean of the "ppage" variable lies within the interval (51.17, 52.21).

## F. Type II error

To calculate the type II error, we need to specify an alternative hypothesis (the mean age is 55 years) and calculate the probability of failing to reject the null hypothesis when the alternative hypothesis is true.


````python
import statsmodels.stats.power as smp

# Calculate the type II error
# Assuming the true mean under the alternative hypothesis is 55
true_mean_alternative = 55
effect_size = (true_mean_alternative - specified_value) / sample_std
power_analysis = smp.TTestPower()
power = power_analysis.solve_power(effect_size=effect_size, nobs=n, alpha=0.02, alternative='two-sided')

type_ii_error = 1 - power
type_ii_error
````

This indicates that there is virtually no probability of failing to reject the null hypothesis when the true mean under the alternative hypothesis is 55 years.

## B. Confidence Interval for the Median using Bootstrapping

````python
# Pick a random sample of size 20
np.random.seed(42)
sample = np.random.choice(dataset[chsooing_variable], size=20, replace=False)

# Perform bootstrapping
n_iterations = 1000
bootstrap_medians = []
for _ in range(n_iterations):
    bootstrap_sample = np.random.choice(sample, size=20, replace=True)
    bootstrap_medians.append(np.median(bootstrap_sample))

# Calculate the 95% confidence interval for the median using the bootstrap method
bootstrap_medians = np.array(bootstrap_medians)
mean_conf_interval = np.percentile(bootstrap_medians, [2.5, 97.5])

# Plot the bootstrap distribution on a dot-plot
plt.figure(figsize=(10, 6))
sns.histplot(bootstrap_medians, bins=30, kde=True, color='lightblue', edgecolor='black')
plt.axvline(x=np.median(sample), color='red', linestyle='--', label='Sample Mean')
plt.axvline(x=mean_conf_interval[0], color='green', linestyle='-.', label='95% CI Lower Bound')
plt.axvline(x=mean_conf_interval[1], color='green', linestyle='-.', label='95% CI Upper Bound')
plt.title(f'Bootstrap Distribution of the Median {chsooing_variable}')
plt.xlabel(f'Median {chsooing_variable}')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()

np.mean(sample), mean_conf_interval
````

  - Median of Weight: 1.012545
  - 95% Confidence Interval: [0.82915, 1.1782]

1. **Range of the Confidence Intervals:**
   - The percentile method gives a much wider confidence interval (0.3769 to 1.7883) compared to the bootstrapping method (0.82915 to 1.1782). This means the percentile method suggests a broader range within which the true median might lie.
   - The narrower confidence interval from the bootstrapping method indicates a more precise estimate of the median weight.

2. **Location of the Median:**
   - The medians calculated by both methods are close but not identical. The percentile method median is 0.9676, while the bootstrapping method median is slightly higher at 1.012545.

3. **Confidence Interval Bounds:**
   - The lower bound of the percentile method (0.3769) is much lower than that of the bootstrapping method (0.82915).
   - The upper bound of the percentile method (1.7883) is much higher than that of the bootstrapping method (1.1782).


- **Percentile Method** is based on the actual data percentiles and does not make any assumptions about the distribution of the data. It is more robust to outliers, which is why it has a wider range. The wider interval suggests higher variability in the data.

- **Bootstrapping Method** involves resampling the data multiple times and computing the median for each resample. This method provides a more empirical estimate based on the observed variability within the sample. The narrower confidence interval suggests less variability within the resampled data.

````python
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Create a dataframe for ANOVA
anova_df = data[['ppage', 'voter_category']].dropna()

# Fit the model
model = ols('ppage ~ C(voter_category)', data=anova_df).fit()

# Perform ANOVA
anova_table = sm.stats.anova_lm(model, typ=2)

anova_table
````

The extremely small p-value indicates that we reject the null hypothesis. This suggests that there are significant differences in the mean age across the different voter categories ("always", "rarely/never", "sporadic").

## B. Hypothesis test for the mean difference

choose two groups, "always" and "sporadic", to perform a hypothesis test for the mean difference of age.

1. **Null Hypothesis (H0)**: There is no difference in mean age between "always" and "sporadic" voters.
2. **Alternative Hypothesis (H1)**: There is a difference in mean age between "always" and "sporadic" voters.

````python
from scipy.stats import ttest_ind

# Divide the dataset into groups based on voter_category
always_group = dataset[dataset['voter_category'] == 'always']['ppage']
rarely_never_group = dataset[dataset['voter_category'] == 'rarely/never']['ppage']
sporadic_group = dataset[dataset['voter_category'] == 'sporadic']['ppage']

# Perform a two-sample t-test
t_statistic, p_value = ttest_ind(always_group, sporadic_group)
t_statistic, p_value
````

- **t-Statistic:** 6.407
- **p-value:** $1.64 \times 10^{-10}$

The p-value is much smaller than the significance level of 0.05. Therefore, we reject the null hypothesis. So, There is a statistically significant difference in the mean ages between "always" and "sporadic" voters. This result is consistent with the ANOVA test, which also indicated significant differences in mean age across different voter categories.
