### a: Scatter Plots
plot the scatter plot of `wage` vs `educ` with and without a log scale for `wage`.

```r
# Load necessary libraries
library(ggplot2)

# Load the dataset
uswages_df <- read.csv("uswages.csv")

# Scatter plot of wage vs. education
p1 <- ggplot(uswages_df, aes(x=educ, y=wage)) +
  geom_point(alpha=0.5) +
  labs(x='Years of Education', y='Wage', title='Wage vs. Education') +
  theme_minimal()

# Scatter plot with log scale for wage
p2 <- ggplot(uswages_df, aes(x=educ, y=log(wage))) +
  geom_point(alpha=0.5) +
  labs(x='Years of Education', y='Log(Wage)', title='Log(Wage) vs. Education') +
  theme_minimal()

# # Plot the graphs side by side
library(gridExtra)
grid.arrange(p1, p2, nrow=2)
```

The scatter plots show the relationship between years of education and wage, with the second plot using a log scale for the wage. The log transformation helps to reduce the skewness of the wage data and may reveal more patterns in the data.

### b: Smoothing Spline Fit

The error indicates that the `x` values (years of education) must be in increasing order for the smoothing spline to work when `s > 0`. So I sort the data by `educ` before fitting the spline.


```r
# Fit a smoothing spline
spline_fit <- smooth.spline(uswages_df$educ, uswages_df$wage)
log_spline_fit <- smooth.spline(uswages_df$educ, log(uswages_df$wage))

# Plot the data and the smoothing spline
p1 <- ggplot(uswages_df, aes(x=educ, y=wage)) +
  geom_point(alpha=0.5) +
  geom_line(data=data.frame(educ=spline_fit$x, wage=spline_fit$y), aes(x=educ, y=wage), color='red') +
  labs(x='Years of Education', y='Wage', title='Wage vs. Education with Smoothing Spline') +
  theme_minimal()

p2 <- ggplot(uswages_df, aes(x=educ, y=log(wage))) +
  geom_point(alpha=0.5) +
  geom_line(data=data.frame(educ=log_spline_fit$x, wage=log_spline_fit$y), aes(x=educ, y=wage), color='red') +
  labs(x='Years of Education', y='Log(Wage)', title='Log(Wage) vs. Education with Smoothing Spline') +
  theme_minimal()

grid.arrange(p1, p2, nrow=2)
```

The plots above show the smoothing spline fits for both the original wage data and the log-transformed wage data.

**The quality of the fit:**
- **Original Wage Scale:** The smoothing spline captures the general trend of increasing wages with more years of education, but there is a noticeable variability in the data that the spline does not fully address.
- **Log Wage Scale:** The spline fit on the log-transformed data appears smoother and reveals more subtle trends in the data, which can be helpful for understanding the relationship between education and wages.

### c: LOWESS Fit
fit the LOWESS method to the data and plot it.

```r
# Fit the LOWESS model
lowess_fit <- loess(wage ~ educ, data=uswages_df, span=0.2)
lowess_log_fit <- loess(log(wage) ~ educ, data=uswages_df, span=0.2)

# Plot the data and the LOWESS fit
p1 <- ggplot(uswages_df, aes(x=educ, y=wage)) +
  geom_point(alpha=0.5) +
  geom_line(data=data.frame(educ=lowess_fit$x, wage=predict(lowess_fit)), aes(x=educ, y=wage), color='blue') +
  geom_line(data=data.frame(educ=spline_fit$x, wage=spline_fit$y), aes(x=educ, y=wage), color='red') +
  labs(x='Years of Education', y='Wage', title='Wage vs. Education with LOWESS Fit') +
  theme_minimal()

p2 <- ggplot(uswages_df, aes(x=educ, y=log(wage))) +
  geom_point(alpha=0.5) +
  geom_line(data=data.frame(educ=lowess_log_fit$x, wage=predict(lowess_log_fit)), aes(x=educ, y=wage), color='blue') +
  geom_line(data=data.frame(educ=log_spline_fit$x, wage=log_spline_fit$y), aes(x=educ, y=wage), color='red') +
  labs(x='Years of Education', y='Log(Wage)', title='Log(Wage) vs. Education with LOWESS Fit') +
  theme_minimal()

grid.arrange(p1, p2, nrow=2);
```

The plots above show the LOWESS fits for both the original wage data and the log-transformed wage data alongside the smoothing spline fits.

**Comparison of LOWESS and Smoothing Spline Fits:**
- **Original Wage Scale:** The LOWESS fit appears to follow the data more closely compared to the smoothing spline. It captures more of the local variations in the wage data, providing a better fit for this dataset.
- **Log Wage Scale:** Similarly, on the log-transformed data, the LOWESS fit adapts to the local changes in the data more effectively than the smoothing spline, suggesting that LOWESS might be more flexible in capturing the underlying patterns.


### Task d: Mean and Median Wages
calculate the mean and median wages for each education level and plot these values.

```r
# Compute mean and median wage for each level of education
mean_wage <- aggregate(wage ~ educ, data=uswages_df, mean)
median_wage <- aggregate(wage ~ educ, data=uswages_df, median)

# Plot mean and median wage vs. education
ggplot() +
  geom_line(data=mean_wage, aes(x=educ, y=wage), color='blue', linetype="dashed", size=1, label='Mean Wage') +
  geom_point(data=mean_wage, aes(x=educ, y=wage), color='blue') +
  geom_line(data=median_wage, aes(x=educ, y=wage), color='red', linetype="dotted", size=1, label='Median Wage') +
  geom_point(data=median_wage, aes(x=educ, y=wage), color='red') +
  labs(x='Years of Education', y='Wage', title='Mean and Median Wage vs. Education') +
  theme_minimal() +
  scale_color_manual(values = c('blue', 'red'), labels = c('Mean Wage', 'Median Wage')) +
  theme(legend.position="bottom")
```

The plot shows how the mean and median wages change with the number of years of education.

**Observations:**
- Both the mean and median wages generally increase with higher levels of education.
- The median wage tends to be less affected by outliers compared to the mean wage, providing a more stable trend.

### e: Quartiles and Median
calculate the 25th percentile (Q1), median (Q2), and 75th percentile (Q3) wages for each level of education and plot these values.

The plot displays the 25th percentile (Q1), median (Q2), and 75th percentile (Q3) wages for each level of education, providing a detailed view of the wage distribution by education level.
