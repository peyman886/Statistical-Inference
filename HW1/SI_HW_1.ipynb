# **3. Identify the Type of Sampling**

a. **Surveying every passenger in five flights on a particularly selected day at an airline company.**
   - **Type of Sampling:** Cluster Sampling
   - **Why:** In this case, the entire population is divided into clusters (each flight can be seen as a cluster), and a few clusters are chosen randomly or by selection for study. On the chosen day, all individuals within the selected clusters (flights) are surveyed.

b. **Putting names of all employees in a bowl and choosing one of the names without looking into the bowl.**
   - **Type of Sampling:** Simple Random Sampling
   - **Why:** Each employee has an equal chance of being selected, which is the defining characteristic of simple random sampling. The selection is made randomly, without any influence by specific characteristics.

c. **Surveying 100 students by getting random samples of 25 seniors, 25 sophomores, 25 juniors, and 25 freshmen.**
   - **Type of Sampling:** Stratified Sampling
   - **Why:** The population of students is divided into strata (groups) based on their year in school. A random sample is then taken from each stratum. This ensures that each subgroup is represented in the sample.

d. **Taking the last phone number of each page in the registration book of a competition to announce the winners of the competition.**
   - **Type of Sampling:** Systematic Sampling
   - **Why:** This approach involves selecting members of the population at regular interval units (in this case, the interval is every page of the registration book). This method assumes that the pages are not ordered in any way that could bias the selection.

## b. Outliers

Outliers are typically defined as values that fall below $Q1 - 1.5 \times IQR$ or above $Q3 + 1.5 \times IQR$, where $IQR$ is the interquartile range $(Q3 - Q1)$.

Let's calculate this for both sets of data.

```r
# Function to detect outliers
detect_outliers <- function(scores) {
  Q1 <- quantile(scores, 0.25)
  Q3 <- quantile(scores, 0.75)
  IQR <- Q3 - Q1
  lower_bound <- Q1 - 1.5 * IQR
  upper_bound <- Q3 + 1.5 * IQR

  outliers <- scores[scores < lower_bound | scores > upper_bound]
  return(outliers)
}

# Detecting outliers for both classes
day_outliers <- detect_outliers(day_scores)
night_outliers <- detect_outliers(night_scores)

cat("Day Class Outliers:", day_outliers, "\n")
cat("Night Class Outliers:", night_outliers, "\n")
```

the analysis reveals:

- **Day Class**: There are no outliers.
- **Night Class**: The outliers are 45.0 and 25.5.

## c. remove outliers?


Not always. The decision to remove outliers should depend on the context of the data and the goal of the analysis. Outliers can sometimes represent valuable extremes or errors in the data. If they are genuine observations, especially in domains like finance or environmental studies, they can provide important insights. However, if they result from errors or significantly skew the data, it might be justified to remove them or to use robust statistical methods that are less sensitive to outliers.

## b. Number of Diamonds in Each Clarity Group

```r
barplot(table(diamonds$clarity), main="Number of Diamonds by Clarity", xlab="Clarity", ylab="Frequency")
```

The bar chart above displays the number of diamonds across different clarity groups. From this visualization, it's evident that some clarity groups are more prevalent in the dataset than others, with specific groups like "SI1" and "VS2" having higher counts compared to others.

## c. Distribution of 'Price'

```r
hist(diamonds$price, main="Histogram of Diamond Prices", xlab="Price", ylab="Frequency")
```

The histogram above illustrates the distribution of diamond prices. This distribution shows a right-skewed pattern, where a majority of the diamonds are concentrated at the lower price range, with fewer diamonds extending towards the higher price range.

The skewness value of the price distribution is approximately 1.62. A skewness value > 0 indicates a right-skewed distribution, which is consistent with the histogram. This means that while most diamonds are priced lower, there are a few diamonds with very high prices that pull the mean to the right, making the distribution asymmetrical.

## d. Distribution of Price Along with 'Clarity'

```r
boxplot(price ~ clarity, data=diamonds, main="Price Distribution by Clarity", xlab="Clarity", ylab="Price")
```

The side-by-side boxplots above illustrate the distribution of diamond prices across different clarity groups. From these plots, we can observe the following:

- The median price varies across different clarity categories, suggesting that clarity does impact price to some extent.
- There are outliers in every clarity group, indicated by the dots outside the whiskers of the boxplots. These outliers represent diamonds that are priced significantly higher or lower than the majority within their clarity category.
- The range of prices (the vertical distance between the bottom and top of each box, plus the whiskers) also varies by clarity, indicating variability in how much diamonds of a particular clarity can be valued.
