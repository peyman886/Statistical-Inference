# Question 6

**Power Curve Analysis for Rejecting $H_0: \mu \geq 28$**

To determine the power curve for rejecting $H_0: \mu \geq 28$, we calculate the power for the following values of $\mu$: $22$, $23$, $24$, $25$, $26$, and $27$.

This R code generates the power curve for the three different scenarios: $n = 50$ and $\alpha = 0.05$, $n = 50$ and $\alpha = 0.01$, and $n = 20$ with $\alpha = 0.05$:

```r
# R code for generating the power curve plot

# Required libraries
library(ggplot2)
library(dplyr)

# Given data
sample_size <- 50
sample_std <- 5.6
alpha <- 0.05
mu_values <- c(22, 23, 24, 25, 26, 27)
population_mean <- 28

# Calculate critical Z-value for alpha = 0.05
critical_z_alpha_005 <- qnorm(1 - alpha)

# Calculate power for each mu value for alpha = 0.05
power_values_alpha_005 <- pnorm((critical_z_alpha_005 - (population_mean - mu_values)) / (sample_std / sqrt(sample_size)))

# New significance level
alpha_new <- 0.01

# Calculate critical Z-value for alpha = 0.01
critical_z_alpha_001 <- qnorm(1 - alpha_new)

# Calculate power for each mu value for alpha = 0.01
power_values_alpha_001 <- pnorm((critical_z_alpha_001 - (population_mean - mu_values)) / (sample_std / sqrt(sample_size)))

# New sample size
sample_size_new <- 20

# Calculate power for each mu value for n = 20 and alpha = 0.05
power_values_sample_size_20 <- pnorm((critical_z_alpha_005 - (population_mean - mu_values)) / (sample_std / sqrt(sample_size_new)))

# Create a data frame for plotting
power_data <- data.frame(
  mu_values = rep(mu_values, 3),
  power = c(power_values_alpha_005, power_values_alpha_001, power_values_sample_size_20),
  condition = rep(c("n = 50, α = 0.05", "n = 50, α = 0.01", "n = 20, α = 0.05"), each = length(mu_values))
)

# Plot power curves
ggplot(power_data, aes(x = mu_values, y = power, color = condition)) +
  geom_line() +
  geom_point() +
  labs(title = "Power Curve for Rejecting H_0: mu_a ≥ 28",
       x = "𝜇 Values",
       y = "Power",
       color = "Condition") +
  theme_minimal() +
  theme(legend.position = "bottom")
```

**Interpret the plot:**

a) Power represents the probability of correctly rejecting a false null hypothesis. In this context, the power values displayed in the graph indicate the likelihood of correctly rejecting $H_0: \mu \geq 28$ for different true population mean values $\mu$. Higher power values suggest a greater ability to detect a true difference from the null hypothesis.

b) The power curve for $α=0.01$ shows the power of the test for each $μ$ value under the stricter significance level. Comparing it to the curve for $α=0.05$, we can observe how reducing the significance level affects the power of the test. Generally, as the significance level decreases, the power of the test also decreases, meaning it becomes less likely to correctly reject a false null hypothesis.


c) The power curve for $n=20$ with $α=0.05$ shows the power of the test for each $μ$ value with a smaller sample size. Comparing it to the curves for $𝑛=50$, we can observe how reducing the sample size affects the power of the test. Generally, as the sample size decreases, the power of the test also decreases, meaning it becomes less likely to correctly reject a false null hypothesis.

**a.**
To generate 20,000 samples of size 60 and calculate the 97% confidence interval for each sample, we performed bootstrap sampling from the dataset of children's heights. After calculating the confidence intervals, we determined the percentage of intervals that include the real mean of the society. Approximately $97.17\%$ of the confidence intervals include the real mean of the society for samples of size 60.

R code:

```r
# Load the dataset
galton <- read.table("Galton.txt", header = TRUE, sep = "\t")

# Function to perform bootstrap sampling
bootstrap_sampling <- function(data, sample_size, n_samples) {
  sample_means <- replicate(n_samples, mean(sample(data, size = sample_size, replace = TRUE)))
  return(sample_means)
}

# Parameters
sample_size <- 60
n_samples <- 20000
confidence_level <- 0.97
alpha <- 1 - confidence_level
z_value <- abs(qnorm(alpha / 2)) # Two-tailed

# Perform bootstrap sampling
sample_means <- bootstrap_sampling(galton$Height, sample_size, n_samples)

# Calculate standard error
std_error <- sd(galton$Height, na.rm = TRUE) / sqrt(sample_size)

# Calculate confidence intervals
ci <- t(sapply(sample_means, function(x) c(x - z_value * std_error, x + z_value * std_error)))

# Calculate real mean of the society
real_mean <- mean(galton$Height, na.rm = TRUE)

# Determine the percentage of intervals including the real mean
percentage_include_mean <- mean(ci[,1] <= real_mean & ci[,2] >= real_mean) * 100
percentage_include_mean
```
